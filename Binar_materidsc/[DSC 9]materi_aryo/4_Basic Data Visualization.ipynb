Based on Description above, we understand that Seaborn is Python data visualization library based on matplotlib. means that Seaborn was created with matplotlib engine. Seaborn provide more simplistic approach when creating statistical plots.

## Barplot
The Syntax plotting Barplot is similar with matplotlib. In this practice section, we will use same dataframe above in the matplotlib section.
```python
sns.barplot(df['City'], df['Sales'])
plt.show()
```

````python
sns.barplot(x= 'City', y = 'Sales', data=df)
plt.show()
````

What is the differences between matplotlib barplot with seaborn ?

While Matplotlib plot is having such basic plotting, Seaborn plotting do have some styling and colour. Even The plot already have axis label.

For More Styling in seaborn, we can use `sns.set_style(style)` for list of style available, please refer to the documentation. For now, set styling as darkgrid :
```python
sns.set_style("darkgrid")
```

````python
sns.set_style("whitegrid")
````

To add title on the searborn figure we can use `plt.title('Your title here')` or also `sns.barplot(x,y).set_title('Your title here')`

## Scatterplot
From _tips_ dataset above, tipping seems interesting, now we want to plot scatterplot to know correlation between `tips` and `total_bill`.

We can use :
```python
sns.scatterplot(tips['total_bill'], tips['tip'])
plt.title('Tips against Total Bill')
plt.show()
```

````python
sns.scatterplot(x = 'total_bill', y='tip',data=tips)
plt.title('Tips against Total Bill')
plt.show()
````

Whoa, seems it's having positive correlation, if you don't understand what is correlation, don't worry, we will learn those term in Module, Statistics for Data Science.

Now we want the Scatterplot also shown day of the week. We can add `hue` to the Scatterplot :

```python
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['day'])
plt.title('Tips against Total Bill')
plt.show()
```


````python
sns.scatterplot(x = 'total_bill', y='tip',data=tips, hue=tips['day'])
plt.title('Tips against Total Bill')
plt.show()
````

instead of putting `df['column_name']` everytime, in seaborn, we can just pass `data=df`, then we can directly call `x='column_name'` instead.

````python
sns.scatterplot(x = 'total_bill', y='tip',data=tips, hue=tips['smoker'])
plt.title('Tips against Total Bill')
plt.show()
````

Now, add column `smoker` as style.
